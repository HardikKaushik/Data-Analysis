```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('darkgrid')
plt.rcParams['font.size']=15
plt.rcParams['figure.figsize']=(10,7)
plt.rcParams['figure.facecolor']='#FFE584'
```

```python
happy_df= pd.read_csv('E:/Data Analysis/Eda with python/World happiness report/world-happiness-report-2021.csv')
```

```python
happy_df.head(10)
```

```python
happy_df.rename(columns={'Ladder score':'Happiness'},inplace=True)
```

```python
happy_df.head()
```

```python
data =['Country name','Regional indicator','Happiness','Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
```

```python
happy_df = happy_df[data].copy()
```

```python
happy_df.head()
```

```python
happy_df.rename(columns={'Regional indicator':'Regional','Logged GDP per capita':'GDP','Freedom to make life choices':'Freedom'},inplace=True)
```

```python
happy_df.head()
```

```python
happy_df.isnull().sum()
```

```python
happy_df.info()
```

```python
#plot between happines and gdp
plt.rcParams['figure.figsize']=(15,7)
plt.title("plot between happiness score and Gdp")
sns.scatterplot(x=happy_df.Happiness,y=happy_df.GDP,hue=happy_df.Regional,s=200)
plt.legend(loc='upper left',fontsize='10')
plt.xlabel("Happiness")
plt.ylabel("GDP per capita")
```

```python
gdp_region=happy_df.groupby('Regional')['GDP'].sum()
```

```python
gdp_region
```

```python
gdp_region.plot.pie(autopct='%1.1f%%')
plt.title('GDP by region')
```

```python
#total countries in the region

total_con=happy_df.groupby('Regional')[['Country name']].count()
total_con
```

```python
#correalation map

cor = happy_df.corr(method='pearson')
f,ax = plt.subplots(figsize=(10,5))
sns.heatmap(cor,mask=np.zeros_like(cor, dtype= np.bool_),
            cmap='Blues',square=True, ax=ax)
```

```python
#coruption in differnt regions

corruption = happy_df.groupby('Regional')[['Perceptions of corruption']].mean()
corruption
```

```python
plt.rcParams['figure.figsize']=(15,7)
name = happy_df['Regional']
price = happy_df['Perceptions of corruption']
plt.bar(name, price)
plt.title("plot of Coruption among Regions")
plt.xlabel("Regions",fontsize=15)
plt.xticks(rotation=30, ha='right')
plt.ylabel("corruption",fontsize=15)
plt.show()
```

```python
#top 10 happy countries
top_10 = happy_df.head(10)

plt.rcParams['figure.figsize']=(15,7)
countries = top_10['Country name']
happiness = top_10['Healthy life expectancy']
plt.bar(countries, happiness)
plt.title("Top 10 happy Countries with countless life expectancy ")
plt.xlabel("Countries",fontsize=15)
plt.xticks(rotation=30, ha='right')
plt.ylabel("life expectancy",fontsize=15)
plt.show()
```

```python
#top 10 Sad countries
top_10 = happy_df.tail(10)

plt.rcParams['figure.figsize']=(15,7)
countries = top_10['Country name']
happiness = top_10['Healthy life expectancy']
plt.bar(countries, happiness)
plt.title("Top 10 Sad Countries with countless life expectancy")
plt.xlabel("Countries",fontsize=15)
plt.xticks(rotation=30, ha='right')
plt.ylabel("life expectancy",fontsize=15)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(15,7)
plt.title("plot between haFreedom to make life Choices vs Happiness Score ")
sns.scatterplot(x=happy_df.Freedom ,y=happy_df.	Happiness,hue=happy_df.Regional,s=200)
plt.legend(loc='upper left',fontsize='10')
plt.xlabel("Freedom to make life Choices")
plt.ylabel("Happiness Score")
```

```python
top_10 = happy_df.tail(10)

plt.rcParams['figure.figsize']=(15,7)
countries = top_10['Country name']
corrupt = top_10['Perceptions of corruption']
plt.bar(countries, corrupt)
plt.title("Top 10 most corrupt contries in the world ")
plt.xlabel("Countries",fontsize=15)
plt.xticks(rotation=30, ha='right')
plt.ylabel("Corruptness",fontsize=15)
plt.show()
```

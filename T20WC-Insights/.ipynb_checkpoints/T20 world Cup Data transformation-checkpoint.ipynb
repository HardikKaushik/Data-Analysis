```python
import pandas as pd
import json
```

```python
with open('E:/Data Analysis/power bi/2_t20wc_analysis/t20_json_files/t20_wc_match_results.json') as f:
    data= json.load(f)
data[0]['matchSummary']
df_data= pd.DataFrame(data[0]['matchSummary'])
df_data.head()
```

```python
df_data.rename({'scorecard':'match_id'},inplace=True,axis=1)
```

```python
dist_matchid ={}

for index, row in df_data.iterrows():
        key1= row['team1'] + ' Vs ' + row['team2']
        key2= row['team2'] + ' Vs ' + row['team1']
        
        dist_matchid[key1] = row['match_id']
        dist_matchid[key2] = row['match_id']

dist_matchid
```

```python
df_data.head()
#exporting match summaryy to csv
df_data.to_csv('E:/Data Analysis/power bi/2_t20wc_analysis/csv_files/match_details.csv',index= False)
```

```python
#batting summaryyyyy
with open('E:/Data Analysis/power bi/2_t20wc_analysis/t20_json_files/t20_wc_batting_summary.json') as f:
    data= json.load(f)

    all_matches=[]
for rec in data:
    all_matches.extend(rec['battingSummary'])

all_matches
df_batting = pd.DataFrame(all_matches)
df_batting
```

```python
df_batting['out/not out']= df_batting.dismissal.apply(lambda x: "out" if len(x)>0 else "not out")
```

```python
df_batting
```

```python
df_batting.drop('dismissal',inplace=True,axis=1)
```

```python
df_batting['match id'] = df_batting['match'].map(dist_matchid)
df_batting
```

```python
df_batting.to_csv('E:/Data Analysis/power bi/2_t20wc_analysis/csv_files/batting.csv',index= False)
```
